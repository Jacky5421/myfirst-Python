```python
import pandas as pd
import numpy as np
```

```python
# read the data into a pandas dataframe


df = pd.read_excel(io="DESIGN REDLINE PROCESS.xlsx", sheet_name="Raw data use")


df.sample(frac=0.05)
```

```python
# df1=df[(df['Building Element affected'].isin(['Truss-S','Blocking-E','Sheathing-E','Plumbing-S'])) & (df['Reason Cat'].isnull()==False)]
# df1.sample(frac=0.1)
df1=df[(df["Issue month"]).isin([4,5,6,7,8,9])&(df["Issue year"]==2019)&(df["Mfg Line"]=="Wall Panel")]
df1.shape
```

```python
# setup the data
X, y = df1['Comment regarding Issue / Redline'], df1['Reason Cat']
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
ytrain
```

```python
# setup a custom preprocessor
import re
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
def preprocess(text):
    # replace one or more white-space characters with a space
    regex = re.compile(r"\s+")                               
    text = regex.sub(' ', text)    
    # lower case
    text = text.lower()                 
    # remove stop words
    sw = stopwords.words('english')
    text = text.split()                                              
    text = ' '.join([w for w in text if w not in sw]) 
    # remove short words
    text = ' '.join([w for w in text.split() if len(w) >= 2])
    # lemmatize
    text = ' '.join([(WordNetLemmatizer()).lemmatize(w) for w in text.split()]) 
    return text
```

```python
# setup the preprocessing->model pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier 
from sklearn.pipeline import Pipeline
clfR = Pipeline(steps=[
    ('pp', TfidfVectorizer(
        #preprocessor=preprocess,
        lowercase=True, stop_words='english', 
        use_idf=False, norm='l2',
        # With L2, it is not the sum of the rows that is equal to 1, but the sum of the squares is equal to 1. The L1 norm will produce a norm where the sum of the values equals 1.ç®€å•å¹³å‡ norm=l1 ï¼Œå¹³æ–¹å¹³å‡ norm=l2
        #None: not normalize
        min_df=1, max_df=1.0, max_features=None, 
        ngram_range=(1, 4))),
        ('mdl',     RandomForestClassifier())
     #   ('mdl',     MultinomialNB())
    #('mdl',     DecisionTreeClassifier(max_depth=None, min_samples_split=2, min_samples_leaf=1, class_weight=None)                 )
    ])
clfM = Pipeline(steps=[
    ('pp', TfidfVectorizer(
        #preprocessor=preprocess,
        lowercase=True, stop_words='english', 
        use_idf=False, norm='l2',
        # With L2, it is not the sum of the rows that is equal to 1, but the sum of the squares is equal to 1. The L1 norm will produce a norm where the sum of the values equals 1.ç®€å•å¹³å‡ norm=l1 ï¼Œå¹³æ–¹å¹³å‡ norm=l2
        #None: not normalize
        min_df=1, max_df=1.0, max_features=None, 
        ngram_range=(1, 4))),
        ('mdl',     MultinomialNB())
    #('mdl',     DecisionTreeClassifier(max_depth=None, min_samples_split=2, min_samples_leaf=1, class_weight=None)                 )
    ])
clfD = Pipeline(steps=[
    ('pp', TfidfVectorizer(
        #preprocessor=preprocess,
        lowercase=True, stop_words='english', 
        use_idf=False, norm='l2',
        # With L2, it is not the sum of the rows that is equal to 1, but the sum of the squares is equal to 1. The L1 norm will produce a norm where the sum of the values equals 1.ç®€å•å¹³å‡ norm=l1 ï¼Œå¹³æ–¹å¹³å‡ norm=l2
        #None: not normalize
        min_df=1, max_df=1.0, max_features=None, 
        ngram_range=(1, 4))),
    ('mdl',     DecisionTreeClassifier(max_depth=None, min_samples_split=2, min_samples_leaf=1, class_weight=None)                 )
    ])
```

```python
# setup grid search
from sklearn.model_selection import GridSearchCV
param_gridR = {
    'mdl__n_estimators':[100,300, 500, 700]
       # 'mdl__alpha':[0.01, 0.1, 0.2, 0.5, 1] # 4 kinds of smoohy
    #'mdl__min_samples_split':[2,3,4,5,6,7,8,9] # 4 kinds of smoohy 
}
param_gridM = {
        'mdl__alpha':[0.01, 0.1, 0.2, 0.5, 1] # 4 kinds of smoohy
    #'mdl__min_samples_split':[2,3,4,5,6,7,8,9] # 4 kinds of smoohy 
}
param_gridD = {
    'mdl__min_samples_split':[2,3,4,5,6,7,8,9] # 4 kinds of smoohy 
}
#combine  param_grid+ clf 
gscvR = GridSearchCV(clfR, param_gridR, iid=False, cv=4, return_train_score=False)
gscvM = GridSearchCV(clfM, param_gridM, iid=False, cv=4, return_train_score=False)
gscvD = GridSearchCV(clfD, param_gridD, iid=False, cv=4, return_train_score=False)
```

```python
# search for best parameters/estimator
gscvR.fit(Xtrain, ytrain)
gscvM.fit(Xtrain, ytrain)
gscvD.fit(Xtrain, ytrain)
#print(gscv.best_estimator_, "\n")
#print(gscv.best_score_, "\n")
print(gscvR.best_params_, "\n")
print(gscvM.best_params_, "\n")
print(gscvD.best_params_, "\n")
#print(gscv.cv_results_, "\n")
```

```python
# evaluate best_estimator_ on test data
ypredR = gscvR.best_estimator_.predict(Xtest)
ypredM = gscvM.best_estimator_.predict(Xtest)
ypredD = gscvD.best_estimator_.predict(Xtest)
from sklearn import metrics
print (metrics.accuracy_score(ytest, ypredR))
print (metrics.accuracy_score(ytest, ypredM))
print (metrics.accuracy_score(ytest, ypredD))
# print (metrics.confusion_matrix(ytest, ypred))
# print (metrics.classification_report(ytest, ypred))
```
